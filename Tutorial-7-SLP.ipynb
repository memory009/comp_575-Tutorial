```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def PlotStatus(X,d,w):
  plt.plot(X[np.where(D[:,0]>=0)[0],1],X[np.where(D[:,0]>=0)[0],2],'.k')
  plt.plot(X[np.where(D[:,0]<0)[0],1],X[np.where(D[:,0]<0)[0],2],'*r')
  q = 1.1
  plt.axis( [ min(X[:,1])-q, max(X[:,1])+q, min(X[:,2])-q, max(X[:,2])+q  ] )
  a= [ min(X[:,1])-q, max(X[:,1])+q]
  a = np.array([a1[0][0,0] for a1 in a])
  b= (w[0,0]-a*w[1,0])/(w[2,0]+0.00001)
  plt.plot(a, b, '--b')
  plt.show()
```

```python
#Problem definition:
X= np.asmatrix([[0,0], [0,1], [1,0], [1,1]]) #inputs
#D= np.asmatrix([0,1,1,1]).T           #target output OR
#D= np.asmatrix([0,0,0,1]).T           #target output AND
D= np.asmatrix([0,1,1,0]).T           #target output XOR
D = 2*D-1 #convert to -1 & +1

xo= np.asmatrix(-np.ones((4,1)))
X=  np.concatenate((xo, X), axis=1) #append fixed input
print(X)
```

```python
  #Initial weights & bias(first component of w)
  ##w=np.asmatrix([-1.5, 1, 1]).T; #2 old exam
  ###w=np.asmatrix([+0.5, -0.5, +1.5]).T;
  # w=np.asmatrix([+10.3, -0.4, +0.8]).T
  # w=np.asmatrix([-2.2,0.5,-1.3]).T
  #w=np.asmatrix([+0.3, -0.3, +0.8]).T;
  w=np.asmatrix([0.5, -1, 1]).T #12
  #w= np.asmatrix([10.12, -55.3, 8.4]).T; #104
  w
```

```python
  #Neuron definition
  lr= 1/2
  f = lambda x: 2*(x>=0)-1 #activation

  PlotStatus(X,D,w)
```

```python
  #Start training
  Y= f(np.matmul(X,w));#output for all 4 patterns in one go [1,-1,1,1]'
  
  learned_ok = np.array_equal(Y, D);
  epoch      = 1;
  updates    = 0;

  while not learned_ok:
    print('\n\n\n********** Epoch: ', epoch, '\n'  );
    #learned_ok

    for i in range(4): #teach 4 patterns

      xi = X[i,:].T #get ith pattern for convenience
      di = D[i,0]   #ith output for convenience

      print('\nProcessing pattern: [{0},{1},{2}] of index {3}, with target output {4}'.format(xi[0,0], xi[1,0], xi[2,0], i, di))
      

      yi= f(np.matmul(w.T,xi))  #estimate output
      print('\nEstimated output  : {0}  (with induced local field ilf= {1})'.format(yi[0,0], np.matmul(w.T,xi)) )

      print('\nCurrent weight    : [{0}, {1}, {2}]'.format(w[0,0], w[1,0], w[2,0])  )
      Dw= lr * (di - yi[0,0]) * xi  #calculate weight change
      print('\nDelta   weight    : [{0}, {1}, {2}]'.format(Dw[0,0], Dw[1,0], Dw[2,0])  )
      w = w + Dw
      print('\nNew     weight    : [{0}, {1}, {2}]'.format(w[0,0], w[1,0], w[2,0]))
      if di == yi[0,0]:
         print(' [NO UPDATE]')
      else:
           updates = updates + 1
      if w[2,0] > 0:
           PlotStatus(X,D,w)    
      #end
      print('\n')

    #for

    learned_ok = np.array_equal( f(np.matmul(X,w)), D )
    #if learned_ok, continue, end
    epoch = epoch + 1
  #while

  print('\n\n\n Your neuron has now learned with {0} weight adjustments in total; Congratulations !\n\n'.format(updates))
```
